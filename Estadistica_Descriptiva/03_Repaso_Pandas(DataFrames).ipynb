## Eliminacion de Columnas

En este caso eliminaremos la columna Unnamed:0 que no nos aporta ningun valor

```python
del df['Unnamed: 0']
#Seleccionamos 5 registros aleatorios de todo el dataframe
df.sample(5)
```

## Seleccionando multiples columnas.

Debemos considerar que pandas es case sensitive es decir hace diferenciacion entre letras mayusculas y minusculas.

```python
df[['TV', 'Radio']].head(5)
```

```python
print('Estamos seleccionando una columna como serie. Notese que el nombre "Radio" ya no esta encerrado entre corchetes')
df['Radio'].sample(3)
```

```python
print('Estamos seleccionando una columna como dataframe. Notese que el nombre "TV" esta encerrado entre corchetes')
df[['TV']].sample(3)
```

## Seleccionar registros especificos (.loc e .iloc)

- Para seleccionar los registros por nombre de indice y columna usaremos ".loc".
- Para seleccionar los registros segun el numero de orden de columnas e indices usaremos ".iloc".

```python
print('Seleccionamos el dato que se encuentra en el indice 121 para TV')
df.loc[121,'TV']
```

```python
print('Seleccionamos el dato que se encuentra en el indice 49 para Radio')
df.iloc[49,1]
```

## Filtrando datos en un DataFrame (Metodo 1 - Masking)

```python
print('Selecionamos todos los registros que tengan ingreso por TV mayor o igual a 100')
mask1 = df['TV'] >= 100
df[mask1]
```

```python
print('Selecionamos todos los registros que tengan ingreso por TV menor a 50 e ingresos por Radio menor a 10')
mask1 = df['TV'] < 50
mask2 = df['Radio'] < 10
df[mask1 & mask2]
```

```python
print('Selecionamos todos los registros que tengan ingreso por Periodico menor a 10 o ingresos por Radio menor a 10')
print('Seleccionaremos solo la columna de Radio y Periodico')
mask1 = df['Newspaper'] < 10
mask2 = df['Radio'] < 10
df[mask1 | mask2][['Newspaper','Radio']]
```

## Filtrando datos en un DataFrame (Metodo 2 - .Query())

```python
radio_less_than_10 = df.query(' Radio < 2')
radio_less_than_10
```

```python
df.query('TV < 10 or Newspaper > 20')
```

```python
df.query('Radio < 10 and Newspaper > 30')
```

## Leemos nuestro segundo archivo

## Agrupando los Datos

```python
print('Necesitamos saber cuantos eventos se han realizado por Ciudad y Deporte')
df.groupby(['City','Sport'])[['Event']].count()
```

```python
print('Necesitamos saber cuantos eventos se han realizado en la ciudad de Turin segun su Deporte')
aux = df.groupby(['City','Sport'])[['Event']].count()
aux.query('City == "Turin"')
```
