# Multiples Graficos en Figuras

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5,0.1)
x2 = np.arange(0,5.0,0.2)
```

```python
plt.figure(1) #este parametro es adicional
plt.subplot(211) #el doble de ancho que de alto y primer grafico, 2 filas y una columna
plt.plot(x1,f(x1), 'ro', x2, f(x2), 'k')

plt.subplot(212) #el doble de ancho que de alto y segundo grafico
plt.plot(x2, f(x2), 'g--')

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])

plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title("Este es el primer titulo")
```

# Texto en los gráficos

```python
mu = 100
sigma = 20
x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor = "g", alpha=0.6)

plt.xlabel("Cociente Intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de CI")
plt.text(20,0.015,r'$\mu = 100, \ \sigma=20$')
plt.grid(True,alpha=0.2)
plt.axis([20,180,0,0.025])
plt.show()
```

```python
plt.subplot(111)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y,linewidth=2.0)

plt.annotate('Maximo Local', xy= (4*np.pi,1), xytext = (15,1.5),
            arrowprops = dict(facecolor = 'black', shrink = 0.05))

plt.ylim(-2,2)
plt.show()
```

```python
mu = 0.5
sd = 0.3

y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()

x = np.arange(len(y))
```

```python
plt.figure(figsize = (10,6))
plt.subplot(221)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True, alpha=0.4)

plt.subplot(222)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logaritmica")
plt.grid(True, alpha=0.4)

plt.subplot(223)
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala Log Simetrico")
plt.grid(True, alpha=0.4)

plt.subplot(224)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala Logistica")
plt.grid(True, alpha=0.4)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.5, wspace = 0.5)
plt.show()
```

# Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S,C = np.sin(x), np.cos(x)
```

```python
plt.plot(x, S, color="blue", linewidth = 1.2, linestyle = "-")
plt.plot(x,C, color="green", linewidth = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks(np.linspace(-1, 1, 9, endpoint = True))
plt.show()
```
