```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Histogramas

```python
df = pd.read_csv(".\datasets\Advertising.csv", sep=",")
del df['Unnamed: 0']
df.head(3)
```

```python
plt.title("Histograma de Publicidad en Radio")
plt.hist(df['Radio'], bins='auto')
plt.xlabel("Los bins o cortes, estan definidos de manera automatica")
plt.show()
```

```python
plt.title("Histograma de Publicidad en Periodico")
plt.hist(df['Newspaper'],bins=5, color='g')
plt.xlabel("Los bins o cortes seran de 5")
plt.show()
```

```python
plt.title("Histograma de Publicidad en TV")
plt.hist(df['TV'],bins=[0,50,150,200,300], color='r')
plt.grid()
plt.xlabel("Asignamos manualmente los rangos de nuestros bins o cortes")
plt.show()
```

```python
df['Sales'].plot.hist(grid=True, bins=10, rwidth=0.95, color="black")
plt.grid(alpha=0.5)
plt.title("Histograma de Ventas")
plt.ylabel("Frecuencia")
plt.show()
```
