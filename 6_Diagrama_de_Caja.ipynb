```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Diagrama de Caja y Bigotes

```python
np.random.seed(19880519)
```

```python
dist1 = 100*np.random.rand(50)
dist2 = 50*np.ones(25)
dist3 = 100+100*np.random.rand(10)
dist4 = -100*np.random.rand(10)

data = np.concatenate((dist1,dist2,dist3,dist4))
```

```python
q75, q25 = np.percentile(data, [75 ,25])
iqr = q75 - q25
print(f'El rango intercuartilico es: {iqr}')
print(f'Se considera outlier si esta por encima de: {(iqr*1.5)+q75}')
print(f'Se considera outlier si esta por debajo de: {q25-(iqr*1.5)}')
```

```python
plt.boxplot(data)
plt.title("Diagrama de Caja")
plt.show()
```

```python
plt.boxplot(data, notch=True)
plt.title("Diagrama de Caja con muescas")
plt.show()
```

```python
green_diamond = dict(markerfacecolor = 'g', marker='D')
plt.boxplot(data, notch=True, flierprops = green_diamond)
plt.title("Diagrama de Caja con muescas")
plt.show()
```

```python
red_square = dict(markerfacecolor = 'r', marker='s')
plt.boxplot(data, notch=True, flierprops = red_square, vert=False)
plt.title("Diagrama de Caja en horizontal")
plt.show()
```

```python
fulldata = [dist1,dist2,dist3]
plt.boxplot(fulldata)
plt.title("Diagrama de Caja")
plt.show()
```
