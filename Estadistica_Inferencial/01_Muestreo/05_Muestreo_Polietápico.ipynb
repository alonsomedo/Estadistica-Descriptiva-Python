# Muestreo Polietápico

Si una vez seleccionada la muestra aleatoria de conglomerados, tomamos de alguna manera una muestra aleatoria de cada uno de ellos, estaremos realizando un **muestreo polietapico**.

## Ejemplo

Para realizar un muestreo polietapico usaremos el dataset de worldcup, podemos elegir una submuestra de 5 jugadores para cada uno de los 4 paises elegidos, obteniendo al final una muestra de tamaño 20 de todos los jugadores de la tabla de datos.

```python
data = pd.read_csv('..\datasets\worldcup.csv', sep=',', index_col=0 )
data.sample(5)
```

Paises Elegidos:

```python
paises_elegidos = np.random.choice(data['Team'], 4, replace=False)
paises_elegidos
```

Ahora elegimos a los 5 jugadores de cada pais:

```python
mask = data['Team'].isin(paises_elegidos)
sample_data = data[mask].groupby('Team').apply(lambda s: s.sample(5))
sample_data
```
